```python
 def train_pre(dic):
    os.mkdir(os.getcwd()+'\\data\\train')
    for i,j in dic.items():
        os.mkdir(os.getcwd()+'\\data\\train\\'+i)
        for c in range(int(0.6*len(j))):
                       shutil.move(os.getcwd()+'\\data\\'+i+'\\'+dic[i][c],os.getcwd()+'\\data\\train\\'+i)
def validation_pre(dic):
    os.mkdir(os.getcwd()+'\\data\\validation')
    for i,j in dic.items():
        os.mkdir(os.getcwd()+'\\data\\validation\\'+i)
        for c in list(np.arange(int(0.6*len(dic[i])),int(len(dic[i])))):
                       shutil.move(os.getcwd()+'\\data\\'+i+'\\'+dic[i][c],os.getcwd()+'\\data\\validation\\'+i)
```
